**Background:**

**Sigmoid:** The Sigmoid function, often denoted as $\sigma$, maps any real-valued number to the range (0, 1), making it especially useful in the output layer of binary classification models. Mathematically, it is represented as:

$\sigma(x)=\frac{1}{1+e^{-x}}$.

Its derivative, essential for the backpropagation algorithm, is given by:

$\sigma^{\prime}(x)=\sigma(x) \cdot(1-\sigma(x))$


**Tanh (Hyperbolic Tangent) Function:** The Tanh function maps any real-valued number to the range (-1, 1). It's similar to the Sigmoid function in shape but has a range that spans from -1 to 1. It is mathematically expressed as:

$\tanh (x)=\frac{2}{1+e^{-2 x}}-1$.

Its derivative, essential for the backpropagation algorithm, is given by:

$\frac{d \tanh (x)}{d x}=1-\tanh ^2(x)$

**Background:**

**ReLU (Rectified Linear Unit):** Defined as $f(x)=\max(0,x)$. It returns $x$ if $x$ is greater than 0, otherwise it returns 0. It's simple and computationally efficient. 

**LeakyReLU:** A variant of ReLU. Instead of having all negative values become zero, it has a small slope for negative values. Given a small parameter $\alpha$, its formula is $f(x)=x \text{ if } x > 0, \text{ else } f(x) = \alpha x$.

**ELU (Exponential Linear Unit):** Tries to make the mean activations closer to zero which speeds up learning. Defined as:

$f(x)=\left\{\begin{array}{ll}
x & \text { if } x>0 \\
\alpha(\exp (x)-1) & \text { otherwise }
\end{array}\right.$


**Swish:** A self-gated activation function defined as $f(x) = x \cdot \sigma(\beta x)$, where $\sigma$ is the sigmoid function and $\beta$ is a learnable parameter.
